```python
#importing required libraries and loading data set
import pandas as pd
anime=pd.read_csv('/content/anime.csv')
anime
```

```python
#looking for missing  values
anime.isnull().sum()
```

```python
# Handling missing values
anime['rating'].fillna(anime['rating'].mean(), inplace=True)
anime.dropna(subset=['genre', 'type'],inplace=True)
anime.isnull().sum()
```

```python
# summarizing key features
anime.describe()
```

```python
# Exploring unique values
print(anime['type'].unique())
print(anime['genre'].unique())
```

```python
# Converting genres to one-hot encoding
anime = anime.join(anime['genre'].str.get_dummies(sep=', '))

# Converting type to numerical representation
anime['type'] = anime['type'].astype('category').cat.codes
```

```python
# Converting 'episodes' column to numeric, handling 'Unknown' values
anime['episodes'] = pd.to_numeric(anime['episodes'], errors='coerce')

# Fill NaN values in 'episodes' after conversion with a suitable value (e.g., mean or median)
anime['episodes'].fillna(anime['episodes'].mean(), inplace=True)

# scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
anime[['episodes', 'members']] = scaler.fit_transform(anime[['episodes', 'members']])

anime
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
features = anime.drop(columns=['anime_id', 'name', 'genre'])
cosine_sim = cosine_similarity(features, features)

# Function to recommend anime
def recommend_anime(title, similarity_matrix=cosine_sim):
    idx = anime[anime['name'] == title].index[0]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 similar anime
    anime_indices = [i[0] for i in sim_scores]
    # Use iloc to select rows by integer position
    return anime.iloc[anime_indices][['name', 'type', 'episodes', 'rating']]

# Example recommendation
print(recommend_anime('Naruto'))
```

```python
#splitting the data
from sklearn.model_selection import train_test_split

# Assume 'rating' as the target variable for simplicity
X_train, X_test, y_train, y_test = train_test_split(features, anime['rating'], test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Using cosine similarity to find nearest neighbors for evaluation
y_pred = []
for idx in X_test.index:
    similar_indices = cosine_similarity([X_test.loc[idx]], X_train)[0].argsort()[-10:]
    recommended_ratings = y_train.iloc[similar_indices].mean()
    y_pred.append(recommended_ratings)

# Convert to binary classification for precision, recall, and F1-score
threshold = 0.5
y_test_binary = (y_test > threshold).astype(int)
y_pred_binary = (pd.Series(y_pred) > threshold).astype(int)

# Compute metrics
precision = precision_score(y_test_binary, y_pred_binary)
recall = recall_score(y_test_binary, y_pred_binary)
f1 = f1_score(y_test_binary, y_pred_binary)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
```
