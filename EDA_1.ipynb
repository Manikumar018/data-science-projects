```python
#importing required libraries and loading the data set
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
hp=pd.read_csv('/content/Cardiotocographic.csv')
hp
```

```python
# Replace null values with the mean of the respective columns
hp.fillna(hp.mean(), inplace=True)
```

```python
#the null values are replaced with the respective columns mean
hp.isnull().sum()
```

```python
hp.info()
#info shows there are no null characters
```

```python
# Defining a function to detect outliers using IQR
def detect_outliers_iqr(data):
  Q1 = data.quantile(0.25)
  Q3 = data.quantile(0.75)
  IQR = Q3 - Q1
  outliers = data[((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))]
  return outliers

# Iterating through numerical columns and detect outliers
for column in hp.select_dtypes(include=np.number).columns:
  outliers = detect_outliers_iqr(hp[column])
  if not outliers.empty:
    print(f"Outliers detected in column '{column}':")
    print(outliers)
    print()
```

```python
# Function to treat outliers by replacing them with the nearest limit
def treat_outliers_iqr(data):
  Q1 = data.quantile(0.25)
  Q3 = data.quantile(0.75)
  IQR = Q3 - Q1
  lower_limit = Q1 - 1.5 * IQR
  upper_limit = Q3 + 1.5 * IQR
  data = np.where(data < lower_limit, lower_limit, np.where(data > upper_limit, upper_limit, data))
  return data

# Iterate through numerical columns and treat outliers
for column in hp.select_dtypes(include=np.number).columns:
    hp[column] = treat_outliers_iqr(hp[column])

# Verify if outliers are still present
for column in hp.select_dtypes(include=np.number).columns:
  outliers = detect_outliers_iqr(hp[column])
  if not outliers.empty:
    print(f"Outliers detected in column '{column}':")
    print(outliers)
    print()
```
