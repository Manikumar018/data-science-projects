```python
#standardising the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
wine_scaled=scaler.fit_transform(wine)
```

```python
#using pca to streamline data and helps in enhancing the performance of analytical models
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)
wine_pca=pca.fit_transform(wine_scaled)
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
print(f"Number of components: {pca.n_components_}")
```

```python
# Scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Variance Explained')
plt.title('Scree Plot')
plt.show()
```

```python
#Applying kmeans to the data
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(wine_scaled)
labels_kmeans=kmeans.labels_
```

```python
# Evaluating clustering performance
from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette_kmeans = silhouette_score(wine_scaled, labels_kmeans)
davies_bouldin_kmeans = davies_bouldin_score(wine_scaled, labels_kmeans)

print(f"Silhouette Score for K-Means: {silhouette_kmeans}")
print(f"Davies-Bouldin Index for K-Means: {davies_bouldin_kmeans}")
```

```python
# Scatter plot for K-Means clusters for standarised data
plt.scatter(wine_scaled[:, 0], wine_scaled[:, 1], c=labels_kmeans, cmap='viridis')
plt.title('K-Means Clustering')
plt.show()
```

```python
# Applying K-Means to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)  # Example: 3 clusters
kmeans_pca.fit(wine_pca)
labels_kmeans_pca = kmeans_pca.labels_

# Evaluating clustering performance
silhouette_kmeans_pca = silhouette_score(wine_pca, labels_kmeans_pca)
davies_bouldin_kmeans_pca = davies_bouldin_score(wine_pca, labels_kmeans_pca)

print(f"Silhouette Score for K-Means with PCA: {silhouette_kmeans_pca}")
print(f"Davies-Bouldin Index for K-Means with PCA: {davies_bouldin_kmeans_pca}")
```

```python
# Scatter plot for K-Means clusters with PCA data
plt.scatter(wine_pca[:, 0], wine_pca[:, 1], c=labels_kmeans_pca, cmap='viridis')
plt.title('K-Means Clustering with PCA')
plt.show()
```

```python
# Comparing silhouette scores and Davies-Bouldin indices
print(f"Silhouette Score without PCA: {silhouette_kmeans}")
print(f"Davies-Bouldin Index without PCA: {davies_bouldin_kmeans}")
print(f"Silhouette Score with PCA: {silhouette_kmeans_pca}")
print(f"Davies-Bouldin Index with PCA: {davies_bouldin_kmeans_pca}")
```
