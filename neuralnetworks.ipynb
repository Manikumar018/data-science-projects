```python
#importing required datasets and loading the dataset
import pandas as pd
alpha=pd.read_csv('/content/Alphabets_data.csv')
alpha
```

```python
#Data normalization
from sklearn.preprocessing import StandardScaler
X=alpha.drop('letter',axis=1)
y=alpha['letter']
scaler=StandardScaler()
X_normalized=scaler.fit_transform(X)
X_normalized
```

```python
#splitting the dataset and training the model
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Initialize and train the MLP classifier
mlp = MLPClassifier(hidden_layer_sizes=(100), max_iter=300, random_state=42)
mlp.fit(X_train, y_train)

# Make predictions on the test set
y_pred = mlp.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
# Defining hyperparameters to test
hidden_layer_sizes_options = [(50,), (100,), (50, 50), (100, 50)]
activation_functions = ['relu', 'tanh', 'logistic']
learning_rate_inits = [0.001, 0.01, 0.1]

# Iterate over different hyperparameter combinations
for hidden_layer_sizes in hidden_layer_sizes_options:
    for activation in activation_functions:
        for learning_rate_init in learning_rate_inits:
            # Initialize and train the MLP classifier with current hyperparameters
            mlp = MLPClassifier(hidden_layer_sizes=hidden_layer_sizes,
                                max_iter=300,
                                random_state=42,
                                activation=activation,
                                learning_rate_init=learning_rate_init)
            mlp.fit(X_train, y_train)

            # Make predictions on the test set
            y_pred = mlp.predict(X_test)

            # Evaluate the model
            accuracy = accuracy_score(y_test, y_pred)
            print(f"Hidden Layers: {hidden_layer_sizes}, Activation: {activation}, Learning Rate: {learning_rate_init}, Accuracy: {accuracy}")
```

```python
from sklearn.model_selection import GridSearchCV
# Define the hyperparameter search space
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['relu', 'tanh', 'logistic'],
    'learning_rate_init': [0.001, 0.01, 0.1]
}
# Performing GridSearchCV
grid_search = GridSearchCV(mlp,param_grid, cv=5, scoring='accuracy', n_jobs=-1) #n_jobs for parallel processing
grid_search.fit(X_train, y_train)

# Printing the best hyperparameters and accuracy
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)

# Evaluating the best model on the test set
best_mlp = grid_search.best_estimator_
y_pred = best_mlp.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy with best hyperparameters: {test_accuracy}")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

# Evaluating the best model on the test set
best_mlp = grid_search.best_estimator_
y_pred = best_mlp.predict(X_test)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred, average='weighted')  # Use 'weighted' for multi-class
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Test Accuracy with best hyperparameters: {test_accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

# Print a classification report for more detailed metrics
print(classification_report(y_test, y_pred))
```

```python
# Assuming 'accuracy' is the variable holding the accuracy of the model with default hyperparameters
# and 'test_accuracy' is the accuracy of the tuned model (from the provided code).

print(f"Default Model Accuracy: {accuracy}")
print(f"Tuned Model Accuracy: {test_accuracy}")


if test_accuracy > accuracy:
  print("The tuned model shows improved performance compared to the default model.")
  print("Hyperparameter tuning led to an increase in accuracy, indicating that the optimized settings better capture the underlying patterns in the data.")
elif test_accuracy < accuracy:
  print("The tuned model shows decreased performance compared to the default model.")
  print("The hyperparameter optimization process may have resulted in a suboptimal configuration. The default hyperparameters might have been better suited for this dataset.")
else:
  print("Both the default and tuned models have similar performance.")
  print("The hyperparameter tuning did not result in significant improvements.")

print("Further analysis of the classification report for precision, recall, and F1-score across different classes can provide a more comprehensive performance comparison and identify specific areas of improvement or decline.")
```
