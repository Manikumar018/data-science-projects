```python
# first column is also data so we remove header

import pandas as pd

retail = pd.read_excel('/content/Online retail.xlsx', header=None)
```

```python
#header name change to items

retail.columns = ['Items']
retail
```

```python
# Split items in each transaction into a list
retail['Items'] = retail['Items'].apply(lambda x: x.split(','))
retail
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Converting the transactions to a list of lists
transactions = retail['Items'].tolist()

# Initializing and fit the transaction encoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

# Converting to a DataFrame
basket = pd.DataFrame(te_ary, columns=te.columns_)

# Display the basket format
print(basket.head())
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Implementing the Apriori algorithm with a specified min_support
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)

# Getting the number of itemsets
num_itemsets = len(frequent_itemsets)

# Generating association rules, passing in num_itemsets
rules = association_rules(frequent_itemsets, num_itemsets=num_itemsets, metric="lift", min_threshold=1, support_only=False)

# Display the rules
print(rules.head())
```
